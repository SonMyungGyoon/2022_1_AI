```python
year = int(input('연도 입력 : '))
if year % 400 == 0 :
  print("%d년은 윤년(Leap year)이다" % year)
elif year % 100 == 0 :
  print("%d년은 평년(Common year)이다" % year)
elif year % 4 == 0 :
  print("%d년은 윤년(Leap year)이다" % year)
else :
  print("%d년은 평년(Common year)이다" % year)
```

```python
year = int(input('연도 입력 : ')) # 입력은 input() 명령 사용. 문자로만 입력됨
if (year % 4 == 0 and year % 100 != 0 or year % 400 ==0) :
  print("%d년은 윤년(Leap year)이다" % year)
else :
  print("%d년은 평년(Common year)이다" % year)
```
